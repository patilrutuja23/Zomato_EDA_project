```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('zomato.csv',encoding='latin-1')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
## In data Missing valvue
df.isnull().sum()
```

```python
[features for features in df.columns if df[features].isnull().sum() > 0]
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.show() 
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country, on = 'Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
country_name = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
# top 3 countries
plt.pie(country_val[:3], labels = country_name[:3], autopct ="%1.2f%%")
plt.show()
```

Max transaction are  from India then US and UK

```python
rating = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
rating
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=rating)
plt.show()
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color',data=rating, palette=['white','red', 'orange','yellow','green', 'green'])
plt.show()
```

```python
# Count Plot
sns.countplot(x='Rating color', hue='Rating color',data=rating, palette=['white','red', 'orange','yellow','green', 'green'])
plt.show()
```
