```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('zomato.csv', encoding='latin-1')
```

```python
df.head()
```

```python
country_code  = pd.read_excel('Country-Code.xlsx')
country_code.head()
```

```python
final_df = pd.merge(df, country_code, on="Country Code", how="left")
final_df.head()
```

```python
final_df.info()
```

```python
final_df.describe()
```

```python
final_df.isnull().sum()
```

```python
# Analysing Rating color for zomato dataset
rating_color = df["Rating color"].value_counts()

plt.figure(figsize=(8,6))
plt.bar(rating_color.index, rating_color.values, color=['Orange', 'White', 'Yellow', 'Green', 'darkgreen', 'Red'])
plt.title("Analysing Rating color for zomato dataset")
plt.xlabel('Colors')
plt.ylabel('Range')
plt.show()
```

```python
final_df[final_df['Has Online delivery'] == 'Yes'].Country.value_counts()
```

```python
# Analysing country have Has Online delivery 
has_online_delivery = final_df.groupby(['Has Online delivery', 'Country']).size().reset_index(name='Count')
has_online_delivery.head()
```

```python
online_delivery = has_online_delivery[has_online_delivery['Has Online delivery'] == 'Yes']

plt.figure(figsize=(8,6))

plt.pie(online_delivery['Count'], labels=online_delivery['Country'], colors=['Orange', 'Green'], autopct='%1.1f%%')
plt.title("Analysing Rating color for zomato dataset")

plt.show()
```

```python
city = final_df.City.value_counts().index
city_value = final_df.City.value_counts().values

plt.figure(figsize=(8,6))

plt.pie(city_value[:6], labels=city[:6], colors=['Orange', 'White', 'Yellow', 'Green', 'darkgreen', 'Red'], autopct='%1.1f%%')
plt.title("Analysing Rating color for Cities in India")
plt.tight_layout()
plt.show()
```
